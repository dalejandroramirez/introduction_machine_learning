```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("Base_datos/diabetes.csv")
data.head()
LenData=len(data)
```

```python
data.dtypes
```

# Visualizacion de datos
    
- Histograma: Da la frecuencia relativa de cada conjunto de datos.

- Grafica de dispersion : Muestra la relacion entre dos features graficandolas en un eje 


```python
f=plt.figure(figsize=(8,4))
ax= f.add_subplot(1,1,1)
data["Age"].hist(ax=ax,density=True)
ax.set_title("Rango de edad del paciente")
ax.set_xlabel("Edad")
ax.set_ylabel("Frecuencia")
f.tight_layout()
```

```python
sns.jointplot(data = data,y="Age",x="Glucose",hue="Outcome")
```

```python
# Ajustar el tamaño de fuente.
plt.rcParams['font.size'] = 15 

# Crear una figura y ajusta su tamaño.
f = plt.figure(figsize=(8,4))

# Crear un subplot o subtrama - al ser una sola figura es 1,1,1.
ax = f.add_subplot(1,1,1)

# Gráfica tus datos usando 'hist'. Pasa el objeto 'ax' a Pandas. Agrega un borde negro con un groso de 2.
data["Outcome"].hist(ax=ax, edgecolor='black', linewidth=2)

# Establece los límites en el eje x.
ax.set_xlim([-0.5, 1.5])

# Establece la frecuencia de tick (valores en el eje x). Tenemos 0 y 1 que corresponden a Sí y No respectivamente.
ax.set_xticks([0, 1])

# Etiquetar xtick labels.
ax.set_xticklabels(["Not", "Yes"])

# Crea el título.
ax.set_title("Diabetes Y/N?")

# Establece la etiqueta del eje X.
ax.set_xlabel("Answer")

# Establece la etiqueta del eje Y.
ax.set_ylabel("Count")

# Establece los límites superior/inferior del eje y.
ax.set_ylim([0, 510])

# hace que las cosas sean bonitas, no es necesario, pero se ajusta al tamaño de la figura.
f.tight_layout()

plt.show()

```

Las clases estan desbalanceadas por lo tanto es necesario mirar que esta ocurriendo, por ejemplo con la edad.


```python
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
binvalues = [20, 25, 30, 35, 40, 85]
bins = binvalues
data["Age"].hist(ax=ax, edgecolor='black', linewidth=2,bins=bins)
ax.set_title("Age range of patients")
ax.set_ylim([0, 510])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
plt.show()
```

Esto muestra que la mayoria de las personas encuestadas tiene un rango de edad entre 20 y 30 años

# Grafica de dispersión con matplotlib


```python
plt.rcParams['font.size'] = 15 
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
ax.scatter(data["Age"], data["BloodPressure"], alpha=0.25)
ax.set_title("Blood pressure vs. age range of patients")
ax.set_ylabel("Blood pressure")
ax.set_xlabel("Age")
f.tight_layout() 
```
